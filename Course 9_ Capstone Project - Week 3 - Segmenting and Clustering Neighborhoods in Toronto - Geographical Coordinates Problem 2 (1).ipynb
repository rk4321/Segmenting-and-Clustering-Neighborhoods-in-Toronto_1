# Segmenting and Clustering Neighborhoods in Toronto

## Problem 2 - Finding the Geographical Coordinates

###### Importing the required libraries

```python
import pandas as pd
import numpy as np
```

###### Reading the .csv file of the Dataframe created in the previous problem (Problem 1)

```python
df=pd.read_csv('DatasetforToronto.csv')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df.head(20)
```

###### Reading the given .csv file with coordinates in a Dataframe

```python
df_geo=pd.read_csv('https://cocl.us/Geospatial_data')
df_geo.head(20)
```

```python
df_geo.shape
```

###### Changing the 'Postal Code' column name in df_geo as 'PostalCode

```python
df_geo=df_geo.rename({'Postal Code':'PostalCode'},axis=1)
df_geo.head(20)
```

###### Combining the two Dataframes

```python
df=df.join(df_geo.set_index('PostalCode'),on='PostalCode')
df.head(20)
```

###### Saving this Dataframe to a .csv file

```python
df.to_csv("TorontoGeoData.csv")
```
